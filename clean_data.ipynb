```python
import pandas as pd

df = pd.read_csv('hamrobazaar_cars.csv')
df.head()
```

```python
df.info()
```

```python
null_count = df.isnull().sum()
print(null_count)
```

```python
df['price_cleaned'] = (df['price']
                    .str.replace('रू.', '', regex=False)
                    .str.replace('Rs.', '', regex=False)
                    .str.replace(',', '', regex=False)
                    .str.strip()
                    .replace('', None)
                    .astype(float)
)
df.head()
```

```python
# Define a list of valid condition values
valid_conditions = ['Brand New', 'Like New', 'Used']
df['condition_cleaned'] = df['condition'].str.replace('|', '', regex=False).str.strip().str.title()

# replacing with unknown if not in valid conditions
df['condition_cleaned'] = df['condition_cleaned'].apply(lambda x: x if x in valid_conditions else 'Unknown')
print(df['condition_cleaned'].value_counts())
```

```python
df['brand_cleaned'] = df['brand'].fillna('Unknown').str.strip().str.title()
```

```python
df['year_cleaned'] = pd.to_numeric(df['year'], errors='coerce')
df=df.dropna(subset=['year_cleaned'])
df['year_cleaned']=df['year_cleaned'].astype(int)
```

```python
df.dtypes
```

```python
# First, split the location column into multiple columns
location_parts = df['location'].str.split(',', expand=True)

# Now assign each part to a new column, handling missing values gracefully
df['area'] = location_parts[0].str.strip()
df['city'] = location_parts[1].str.strip()
df['district'] = location_parts[2].str.strip()
df['province'] = location_parts[3].str.strip() if location_parts.shape[1] > 3 else None
```

```python
null_count = df.isnull().sum()
print(null_count)
```

```python
import numpy as np
import re

def parse_posted_time(text):
    text = text.lower()
    if "second" in text:
        return 1, "1 minute"
    elif "minute" in text:
        num = re.findall(r'\d+', text)
        minutes = int(num[0]) if num else 1
        return minutes, f"{minutes} minute{'s' if minutes > 1 else ''}"
    elif "hour" in text:
        num = re.findall(r'\d+', text)
        hours = int(num[0]) if num else 1
        return hours * 60, f"{hours} hour{'s' if hours > 1 else ''}"
    elif "day" in text:
        num = re.findall(r'\d+', text)
        days = int(num[0]) if num else 1
        return days * 1440, f"{days} day{'s' if days > 1 else ''}"
    else:
        return np.nan, "Unknown"

# Apply to DataFrame
df[["posted_minutes", "posted_display"]] = df["posted"].apply(
    lambda x: pd.Series(parse_posted_time(x))
)
```

```python
from datetime import datetime, timedelta
import numpy as np
scrape_time = datetime.now()
df['scraped_at'] = scrape_time
```

```python
def parse_posted_time(posted_str, reference_time):
    try:
        posted_str = posted_str.lower().strip()
        
        if "min" in posted_str:
            minutes = int(posted_str.split()[0])
            return reference_time - timedelta(minutes=minutes)
        
        elif "hour" in posted_str:
            hours = int(posted_str.split()[0])
            return reference_time - timedelta(hours=hours)
        
        elif "day" in posted_str:
            days = int(posted_str.split()[0])
            return reference_time - timedelta(days=days)
        
        else:
            return np.nan  
    except:
        return np.nan

# Apply to your dataframe
df['posted_datetime'] = df.apply(lambda row: parse_posted_time(row['posted'], row['scraped_at']), axis=1)
# After you created 'posted_datetime', extract just the date
df['posted_date'] = df['posted_datetime'].dt.date
```

```python
def extract_seller_info(raw_name):
    match = re.search(r"(.+?)(\d+)\s*Ads", raw_name)
    if match:
        seller_name = match.group(1).strip()
        seller_ads = int(match.group(2))
    else:
        seller_name = raw_name.strip()
        seller_ads = 1
    return pd.Series([seller_name, seller_ads])

df[['seller_name_clean', 'seller_ads']] = df['seller_name'].apply(extract_seller_info)
```

```python
df.drop(columns=['price','year','brand','condition','posted','link','image','location','seller_name','scraped_at'], inplace=True)
df.drop(columns=['contact_info'],inplace=True)
df.drop(columns=['seller_total_ads'], inplace=True)
```

```python
exchange_keywords = ['exchange', 'swap', 'trade', 'part exchange']

def has_exchange_option(text):
    if not isinstance(text, str):
        return False
    text = text.lower()
    return any(keyword in text for keyword in exchange_keywords)

# Apply to the 'title' column and create a new boolean column 'exchange_option'
df['exchange_option'] = df['title'].apply(has_exchange_option)
```

```python
import re

def extract_kilometers(text):
    if not isinstance(text, str):
        return None
    # Look for patterns like 50000 km, 50,000 kms, 50k kms, etc.
    match = re.search(r'(\d{1,3}(?:,\d{3})*|\d+)(\s*k?m?s?)', text.lower())
    if match:
        # Clean the number, remove commas
        km_str = match.group(1).replace(',', '')
        try:
            km_value = int(km_str)
            return km_value
        except:
            return None
    return None

# Apply to your text column, e.g. 'title' or 'description' or combined fields
df['kilometers'] = df['title'].apply(extract_kilometers)
```
