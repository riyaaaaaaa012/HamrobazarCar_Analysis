```python
import pandas as pd

df = pd.read_csv('hamrobazaar_cars.csv')
df.head()
```

```python
df.info()
```

```python
null_count = df.isnull().sum()
print(null_count)
```

```python
df['price_cleaned'] = (df['price']
                    .str.replace('रू.', '', regex=False)
                    .str.replace('Rs.', '', regex=False)
                    .str.replace(',', '', regex=False)
                    .str.strip()
                    .replace('', None)
                    .astype(float)
)
df.head()
```

```python
# Define a list of valid condition values
valid_conditions = ['Brand New', 'Like New', 'Used']
df['condition_cleaned'] = df['condition'].str.replace('|', '', regex=False).str.strip().str.title()

# replacing with unknown if not in valid conditions
df['condition_cleaned'] = df['condition_cleaned'].apply(lambda x: x if x in valid_conditions else 'Unknown')
print(df['condition_cleaned'].value_counts())
```

```python
df['brand_cleaned'] = df['brand'].fillna('Unknown').str.strip().str.title()
```

```python
df['year_cleaned'] = pd.to_numeric(df['year'], errors='coerce')
df=df.dropna(subset=['year_cleaned'])
df['year_cleaned']=df['year_cleaned'].astype(int)
```

```python
df.dtypes
```

```python
df['location'] = df['location'].fillna('Unknown')
```

```python
null_count = df.isnull().sum()
print(null_count)
```

```python
import numpy as np
import re

def parse_posted_time(text):
    text = text.lower()
    if "second" in text:
        return 1, "1 minute"
    elif "minute" in text:
        num = re.findall(r'\d+', text)
        minutes = int(num[0]) if num else 1
        return minutes, f"{minutes} minute{'s' if minutes > 1 else ''}"
    elif "hour" in text:
        num = re.findall(r'\d+', text)
        hours = int(num[0]) if num else 1
        return hours * 60, f"{hours} hour{'s' if hours > 1 else ''}"
    elif "day" in text:
        num = re.findall(r'\d+', text)
        days = int(num[0]) if num else 1
        return days * 1440, f"{days} day{'s' if days > 1 else ''}"
    else:
        return np.nan, "Unknown"

# Apply to DataFrame
df[["posted_minutes", "posted_display"]] = df["posted"].apply(
    lambda x: pd.Series(parse_posted_time(x))
)
```

```python
df.head()
```

```python
df.drop(columns=['price','year','brand','condition','posted','link','image'], inplace=True)
```

```python
df = df.rename(columns={
    'price_cleaned': 'price',
    'condition_cleaned': 'condition',
    'year_cleaned': 'year',
    'brand_cleaned': 'brand'


})
```
